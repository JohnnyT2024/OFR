## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df62['FW'], df62['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue')
plt.scatter(df63['FW'], df63['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='red')
plt.scatter(df64['FW'], df64['AREA_COUNTS'], label='+ve MICROWAVE_POS', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df62, df63, df64)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df62_binned = bin_and_sum(df62)
df63_binned = bin_and_sum(df63)
df64_binned = bin_and_sum(df64)

# Plot the binned sums versus 'FW'
plt.plot(df62_binned['FW_bin'], df62_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df63_binned['FW_bin'], df63_binned['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='red', marker='o')
plt.plot(df64_binned['FW_bin'], df64_binned['AREA_COUNTS'], label='+ve MICROWAVE_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df62_binned = bin_and_sum(df62)
df63_binned = bin_and_sum(df63)
df64_binned = bin_and_sum(df64)

# Applying the scaling factors
df62_binned['AREA_COUNTS'] *= 1.000
df63_binned['AREA_COUNTS'] *= 0.133
df64_binned['AREA_COUNTS'] *= 0.000

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df62_binned['FW_bin'], df62_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df63_binned['FW_bin'], df63_binned['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='red', marker='o')
plt.plot(df64_binned['FW_bin'], df64_binned['AREA_COUNTS'], label='+ve MICROWAVE_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df71, df72, df73)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df71_binned = bin_and_sum(df71)
df72_binned = bin_and_sum(df72)
df73_binned = bin_and_sum(df73)

# Plot the binned sums versus 'FW'
plt.plot(df71_binned['FW_bin'], df71_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df72_binned['FW_bin'], df72_binned['AREA_COUNTS'], label='+ve ABRASION_POS', color='red', marker='o')
plt.plot(df73_binned['FW_bin'], df73_binned['AREA_COUNTS'], label='+ve ULTRASONIC_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df71_binned = bin_and_sum(df71)
df72_binned = bin_and_sum(df72)
df73_binned = bin_and_sum(df73)

# Applying the scaling factors
df71_binned['AREA_COUNTS'] *= 1.000
df72_binned['AREA_COUNTS'] *= 1.305
df73_binned['AREA_COUNTS'] *= 0.832

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df71_binned['FW_bin'], df71_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df72_binned['FW_bin'], df72_binned['AREA_COUNTS'], label='+ve ABRASION_POS', color='red', marker='o')
plt.plot(df73_binned['FW_bin'], df73_binned['AREA_COUNTS'], label='+ve ULTRASONIC_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

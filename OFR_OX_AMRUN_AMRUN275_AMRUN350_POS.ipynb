## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df68['FW'], df68['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue')
plt.scatter(df69['FW'], df69['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red')
plt.scatter(df70['FW'], df70['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df68, df69, df70)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df68_binned = bin_and_sum(df68)
df69_binned = bin_and_sum(df69)
df70_binned = bin_and_sum(df70)

# Plot the binned sums versus 'FW'
plt.plot(df68_binned['FW_bin'], df68_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df69_binned['FW_bin'], df69_binned['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red', marker='o')
plt.plot(df70_binned['FW_bin'], df70_binned['AREA_COUNTS'], label='+ve AMRUN 350oC', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df68_binned = bin_and_sum(df68)
df69_binned = bin_and_sum(df69)
df70_binned = bin_and_sum(df70)

# Applying the scaling factors
df68_binned['AREA_COUNTS'] *= 1.000
df69_binned['AREA_COUNTS'] *= 0.683
df70_binned['AREA_COUNTS'] *= 0.133

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df68_binned['FW_bin'], df68_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df69_binned['FW_bin'], df69_binned['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red', marker='o')
plt.plot(df70_binned['FW_bin'], df70_binned['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df42['FW'], df42['AREA_COUNTS'], label='-ve AMRUN', color='blue')
plt.scatter(df43['FW'], df43['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red')
plt.scatter(df44['FW'], df44['AREA_COUNTS'], label='-ve MICROWAVE', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df42, df43, df44)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df42_binned = bin_and_sum(df42)
df43_binned = bin_and_sum(df43)
df44_binned = bin_and_sum(df44)

# Plot the binned sums versus 'FW'
plt.plot(df42_binned['FW_bin'], df42_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df43_binned['FW_bin'], df43_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red', marker='o')
plt.plot(df44_binned['FW_bin'], df44_binned['AREA_COUNTS'], label='-ve MICROWAVE', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 50 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df42_binned = bin_and_sum(df42)
df43_binned = bin_and_sum(df43)
df44_binned = bin_and_sum(df44)

# Applying the scaling factors
df42_binned['AREA_COUNTS'] *= 1.000
df43_binned['AREA_COUNTS'] *= 0.133
df44_binned['AREA_COUNTS'] *= 0.000

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df42_binned['FW_bin'], df42_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df43_binned['FW_bin'], df43_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red', marker='o')
plt.plot(df44_binned['FW_bin'], df44_binned['AREA_COUNTS'], label='-ve MICROWAVE', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

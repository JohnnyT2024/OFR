## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df54['FW'], df54['AREA_COUNTS'], label='-ve AMRUN POS', color='blue')
plt.scatter(df55['FW'], df55['AREA_COUNTS'], label='-ve AMRUN 350oC POS', color='red')
plt.scatter(df56['FW'], df56['AREA_COUNTS'], label='-ve MICROWAVE POS', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df54, df55, df56)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df54_binned = bin_and_sum(df54)
df55_binned = bin_and_sum(df55)
df56_binned = bin_and_sum(df56)

# Plot the binned sums versus 'FW'
plt.plot(df54_binned['FW_bin'], df54_binned['AREA_COUNTS'], label='-ve AMRUN POS', color='blue', marker='o')
plt.plot(df55_binned['FW_bin'], df55_binned['AREA_COUNTS'], label='-ve AMRUN 350oC POS', color='red', marker='o')
plt.plot(df56_binned['FW_bin'], df56_binned['AREA_COUNTS'], label='-ve MICROWAVE POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df54_binned = bin_and_sum(df54)
df55_binned = bin_and_sum(df55)
df56_binned = bin_and_sum(df56)

# Applying the scaling factors
df54_binned['AREA_COUNTS'] *= 1.000
df55_binned['AREA_COUNTS'] *= 0.4995
df56_binned['AREA_COUNTS'] *= 0.0518

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df54_binned['FW_bin'], df54_binned['AREA_COUNTS'], label='-ve AMRUN POS', color='blue', marker='o')
plt.plot(df55_binned['FW_bin'], df55_binned['AREA_COUNTS'], label='-ve AMRUN 350oC POS', color='red', marker='o')
plt.plot(df56_binned['FW_bin'], df56_binned['AREA_COUNTS'], label='-ve MICROWAVE POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df31['FW'], df31['AREA_COUNTS'], label='-ve AMRUN', color='blue')
plt.scatter(df32['FW'], df32['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red')
plt.scatter(df33['FW'], df33['AREA_COUNTS'], label='-ve MICROWAVE', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df31, df32, df33)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df31_binned = bin_and_sum(df31)
df32_binned = bin_and_sum(df32)
df33_binned = bin_and_sum(df33)

# Plot the binned sums versus 'FW'
plt.plot(df31_binned['FW_bin'], df31_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df32_binned['FW_bin'], df32_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red', marker='o')
plt.plot(df33_binned['FW_bin'], df33_binned['AREA_COUNTS'], label='-ve MICROWAVE', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 50 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df31_binned = bin_and_sum(df31)
df32_binned = bin_and_sum(df32)
df33_binned = bin_and_sum(df33)

# Applying the scaling factors
df31_binned['AREA_COUNTS'] *= 1.000
df32_binned['AREA_COUNTS'] *= 0.4995
df33_binned['AREA_COUNTS'] *= 0.0518

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df31_binned['FW_bin'], df31_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df32_binned['FW_bin'], df32_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='red', marker='o')
plt.plot(df33_binned['FW_bin'], df33_binned['AREA_COUNTS'], label='-ve MICROWAVE', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

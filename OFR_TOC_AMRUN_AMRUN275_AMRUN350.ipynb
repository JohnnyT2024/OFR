## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df36['FW'], df36['AREA_COUNTS'], label='-ve AMRUN', color='blue')
plt.scatter(df37['FW'], df37['AREA_COUNTS'], label='-ve AMRUN 275oC', color='red')
plt.scatter(df38['FW'], df38['AREA_COUNTS'], label='-ve AMRUN 350oC', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df31, df32, df33)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df36_binned = bin_and_sum(df36)
df37_binned = bin_and_sum(df37)
df38_binned = bin_and_sum(df38)

# Plot the binned sums versus 'FW'
plt.plot(df36_binned['FW_bin'], df36_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df37_binned['FW_bin'], df37_binned['AREA_COUNTS'], label='-ve AMRUN 275oC', color='red', marker='o')
plt.plot(df38_binned['FW_bin'], df38_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df36_binned = bin_and_sum(df36)
df37_binned = bin_and_sum(df37)
df38_binned = bin_and_sum(df38)

# Applying the scaling factors
df36_binned['AREA_COUNTS'] *= 1.000
df37_binned['AREA_COUNTS'] *= 0.8995
df38_binned['AREA_COUNTS'] *= 0.4995

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df36_binned['FW_bin'], df36_binned['AREA_COUNTS'], label='-ve AMRUN', color='blue', marker='o')
plt.plot(df37_binned['FW_bin'], df37_binned['AREA_COUNTS'], label='-ve AMRUN 275oC', color='red', marker='o')
plt.plot(df38_binned['FW_bin'], df38_binned['AREA_COUNTS'], label='-ve AMRUN 350oC', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

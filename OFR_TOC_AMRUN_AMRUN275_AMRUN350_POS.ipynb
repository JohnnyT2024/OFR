## Exploratory Data Analysis

```python
# Plotting
plt.scatter(df59['FW'], df59['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue')
plt.scatter(df60['FW'], df60['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red')
plt.scatter(df61['FW'], df61['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='green')


# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Area Counts')
plt.title('Combined Scatter Plot of Area Counts vs FW')
plt.legend()

# Show plot
plt.show()
```

```python
# Dataframes (df59, df60, df61)


# Bin the FW values in increments of 25 and sum the AREA_COUNTS within each bin for each dataset
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

df59_binned = bin_and_sum(df59)
df60_binned = bin_and_sum(df60)
df61_binned = bin_and_sum(df61)

# Plot the binned sums versus 'FW'
plt.plot(df59_binned['FW_bin'], df59_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df60_binned['FW_bin'], df60_binned['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red', marker='o')
plt.plot(df61_binned['FW_bin'], df61_binned['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```

### Weighted bins of 25 Plots

```python
# Function to bin the FW values and sum AREA_COUNTS within each bin
def bin_and_sum(df, bin_size=25):
    bins = np.arange(df['FW'].min(), df['FW'].max() + bin_size, bin_size)
    df['FW_bin'] = pd.cut(df['FW'], bins=bins, labels=bins[:-1])
    binned_data = df.groupby('FW_bin')['AREA_COUNTS'].sum().reset_index()
    return binned_data

# Binning the dataframes
df59_binned = bin_and_sum(df59)
df60_binned = bin_and_sum(df60)
df61_binned = bin_and_sum(df61)

# Applying the scaling factors
df59_binned['AREA_COUNTS'] *= 1.000
df60_binned['AREA_COUNTS'] *= 0.8995
df60_binned['AREA_COUNTS'] *= 0.4995

# Plotting the binned and scaled sums versus 'FW'
plt.plot(df59_binned['FW_bin'], df59_binned['AREA_COUNTS'], label='+ve AMRUN_POS', color='blue', marker='o')
plt.plot(df60_binned['FW_bin'], df60_binned['AREA_COUNTS'], label='+ve AMRUN 275oC_POS', color='red', marker='o')
plt.plot(df61_binned['FW_bin'], df61_binned['AREA_COUNTS'], label='+ve AMRUN 350oC_POS', color='green', marker='o')

# Adding labels and legend
plt.xlabel('FW')
plt.ylabel('Sum of Area Counts')
plt.title('Sum of Area Counts in Bins of 25 vs FW')
plt.legend()

# Show plot
plt.show()
```
